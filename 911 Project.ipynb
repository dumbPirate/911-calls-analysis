```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('location of 911_csv')
df.head()
```

```python
df.info()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
df['title'].nunique()
```

```python
y = []
def tere():
    for x in df['title']:
        y.append(x.split(':')[0])
        
tere()
df['Reason'] = pd.Series(y)
```

```python
df['Reasons 2'] = pd.Series(map(lambda x: x.split(':')[0],df['title']))
```

```python
df['Reasons 3'] = df['title'].apply(lambda x:x.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(df['Reason'])
```

```python
type(df['timeStamp'][0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda x:x.hour)
```

```python
df['Month'] = df['timeStamp'].apply(lambda x:x.month)
```

```python
df['dayofweek'] = df['timeStamp'].apply(lambda x:x.dayofweek)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['dayofweek'] = df['dayofweek'].map(dmap)
df.head()
```

```python
sns.countplot(df['dayofweek'], hue = df['Reason'])
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
sns.countplot(df['Month'], hue = df['Reason'])
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

```python
rt = df.groupby('Month').count()
rt
```

```python
rt['twp'].plot()
```

```python
sns.lmplot('Month', 'twp', rt)
```

```python
df['Date'] = df['timeStamp'].apply(lambda x:x.date())
df
```
